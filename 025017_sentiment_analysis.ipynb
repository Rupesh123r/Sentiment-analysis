```python
df2=pd.merge(df,cf,on="Date")
```

```python
df2.head()
```

```python
df2['review'] = df2[df2.columns[3:27]].apply(lambda x: ','.join(x.dropna().astype(str)), axis=1)
```

```python
df2.head()
```

```python
review = df2['review'].to_string(header=False,index=False)
```

```python
cleaned_review = re.sub(re.compile('<.*?>'), '', review) #removing HTML tags
cleaned_review = re.sub('[^A-Za-z0-9]+', ' ', review) #taking only words

print(cleaned_review[0:10])
```

```python
cleaned_review = cleaned_review.replace('b', '')
```

```python
# Lowercase

cleaned_review = cleaned_review.lower()

print(cleaned_review[0:10])
```

```python
# Tokenization

# import nltk
nltk.download('punkt')
tokens = nltk.word_tokenize(cleaned_review)

print(tokens[0:10])
```

```python
# Stop words removal

nltk.download('stopwords')
stop_words = stopwords.words('english')
filtered_review = [word for word in tokens if word not in stop_words] # removing stop words

print(filtered_review[0:10])
```

```python
# Stemming

stemmer = PorterStemmer()
stemmed_review = [stemmer.stem(word) for word in filtered_review]

print(stemmed_review[0:10])
```

```python
# Lemmatization
nltk.download('wordnet')
nltk.download('omw-1.4')

lemmatizer = WordNetLemmatizer()
lemm_review = [lemmatizer.lemmatize(word) for word in filtered_review]

print(lemm_review[0:10])
```
