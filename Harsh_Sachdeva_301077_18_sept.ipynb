<h3>Question 2: Clean all the hashtags, urls, mentions and emoticons from the Tweets. Also, remove
duplicates and blank rows</h3>

```python
result.isnull().sum()
```

```python
result = result.drop_duplicates()
```

```python
# Converting the column to string

df = result['2: Tweet'].to_string(header=False,index=False)
```

```python
# Data Cleaning

df_review = re.sub(re.compile('<.*?>'), '', df) #removing HTML tags
df_review = re.sub('[^A-Za-z0-9]+', ' ', df_review) #taking only words

print(df_review)
```

<h3>Question 3: Preprocess the data to (a) remove stopwords and unnecessary characters, (b) lower
the case, (c) remove punctuation, numbers, white spaces and unnecessary patterns,
and (d) apply stemming or lemmatization</h3>

```python
# Lowercase

df_review = df_review.lower()

print(df_review)
```

```python
# Tokenization

import nltk

from nltk.tokenize import word_tokenize

tokens = nltk.word_tokenize(df_review)
```

```python
# Stop words removal

#nltk.download('stopwords')
from nltk.corpus import stopwords

stop_words = stopwords.words('english')

filtered_review = [word for word in tokens if word not in stop_words] # removing stop words

print(filtered_review[0:5])
```

```python
# Stemming

from nltk.stem import PorterStemmer

stemmer = PorterStemmer()

stemmed_review = [stemmer.stem(word) for word in filtered_review]

print(stemmed_review[0:5])
```

```python
# Lemmatization
import nltk
#nltk.download('wordnet')
#nltk.download('omw-1.4')
from nltk.stem import WordNetLemmatizer
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

lemm_review = [lemmatizer.lemmatize(word) for word in filtered_review]

print(lemm_review[0:5])
```
